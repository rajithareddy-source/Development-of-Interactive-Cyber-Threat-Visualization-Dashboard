```python
import pandas as pd
import sqlite3

df = pd.read_csv("cyber_threat_detection_dummy_dataset.csv")
conn = sqlite3.connect("cyber_threat.db")
df.to_sql("cyber_incidents", conn, if_exists="replace", index=False)
```

```python
def run_query(query):
    return pd.read_sql_query(query, conn)
```

```python
run_query("SELECT * FROM cyber_incidents\nLIMIT 10;")
```

```python
run_query("SELECT COUNT(*) AS total_incidents\nFROM cyber_incidents;")
```

```python
run_query("SELECT attack_type, COUNT(*) AS count\nFROM cyber_incidents\nGROUP BY attack_type\nORDER BY count DESC;")
```

```python
run_query("SELECT severity_level, COUNT(*) AS total\nFROM cyber_incidents\nGROUP BY severity_level;")
```

```python
run_query("SELECT is_malicious, COUNT(*) AS total\nFROM cyber_incidents\nGROUP BY is_malicious;")
```

```python
run_query("SELECT attack_type, COUNT(*) AS malicious_count\nFROM cyber_incidents\nWHERE is_malicious = 1\nGROUP BY attack_type\nORDER BY malicious_count DESC;")
```

```python
run_query("SELECT source_country, COUNT(*) AS attacks\nFROM cyber_incidents\nGROUP BY source_country\nORDER BY attacks DESC\nLIMIT 5;")
```

```python
run_query("SELECT destination_port, COUNT(*) AS hits\nFROM cyber_incidents\nGROUP BY destination_port\nORDER BY hits DESC;")
```

```python
run_query("SELECT attack_type, AVG(packet_count) AS avg_packets\nFROM cyber_incidents\nGROUP BY attack_type\nORDER BY avg_packets DESC;")
```

```python
run_query("SELECT severity_level, ROUND(AVG(response_time_ms),2) AS avg_response_time\nFROM cyber_incidents\nGROUP BY severity_level;")
```

```python
run_query("SELECT *\nFROM cyber_incidents\nWHERE severity_level IN ('High', 'Critical')\nAND attack_status = 'Successful';")
```

```python
run_query("SELECT DATE(timestamp) AS attack_date, COUNT(*) AS total_attacks\nFROM cyber_incidents\nGROUP BY attack_date\nORDER BY attack_date;")
```
